# SELECT cmd

- '*'ASPIRIN is used to selecting all items from the table

- ; semi-colon is used identify the line end

- SELECT * FROM employee_demographics - selecting items from the table
- syntax - SELECT * TABLE_NAME 

- SELECT * FROM parks_and_recreation.employee_demographics - select items from a table on other database

# SELECT PARTICULAR / MULTIPLE ITEMS

- SYNTAX - SELECT ITEM FROM TABLE_NAME;

- EG : SELECT first_name FROM employee_demographics;

- SYNTAX - SELECT ITEM,n(ITEMS) FROM TABLE_NAME;

- EG : SELECT first_name,last_name FROM employee_demographics;

# OPERATIONS IN SELECT -  PEMDAS

Order of exections of operation (PEMDAS)

- p- paranthesis
- e- exponents
- m- multiple
- d- divide
- a- add
- s- sum

EG : SELECT age+10*10 FROM employee_demographics;
- operations must executing based on the order

# DISTINCT KEYWORD IN SELECT 

It show the unique values when the tables same value of many then it show a one value .

EG : SELECT distinct gender FROM employee_demographics; ( it show male , female only one time)
