## metrics and KPI (Key performance indicator)

Metrics are measurements of business performance.

Examples:

Total Sales = ₹50,000
Number of Customers = 200

What are KPIs?

KPIs are specific metrics that are important for achieving a business goal.

Example:

Goal: Increase restaurant revenue.

Example of KPI and Metrics

Suppose your restaurant has:

1,000 customers/month
Revenue = ₹2,00,000
Average Order Value = ₹200

These are metrics.

If the owner's goal is to reach ₹3,00,000 revenue, then:

Revenue becomes a KPI.
Average Order Value may also become a KPI because increasing it helps reach the goal.
